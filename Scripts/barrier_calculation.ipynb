```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
down = {}
down['Ba'] = -0.0096751813941247
down['Ti'] = 0.4751970394474812
down['O1'] = 0.0143399835559606 
down['O2'] = 0.5050690791953424 
down['O3'] = 0.5050690791953424 
tmp = down['Ba']
for atom in down:
    down[atom] -= tmp
```

```python
down
```

```python
up = {}
up['Ba'] = 0.0096751721480457
up['Ti'] = 0.5248029530407297
up['O1'] = -0.0143399874138379
up['O2'] = 0.4949309311125303
up['O3'] = 0.4949309311125303
tmp = up['Ba']
for atom in up:
    up[atom] -= tmp
```

```python
up
```

```python
shift = []
for atom in up: 
    shift.append(up[atom] - down[atom])
shift = np.asarray(shift)
qp1 = shift / 2
```

```python
np.abs(sum(np.asarray(list(down.values()) + 2 * qp1) - np.asarray(list(up.values())))) < 1E-16 
```

```python
npoints = 10
qps = []
energies = []
with open('BTO_barrier.txt') as f: 
    for line in f.readlines():
        step = int(line.split()[0])
        qp = step / npoints - 1.0
        energy = float(line.split()[1])
        qps.append(qp)
        energies.append(energy)
qp = np.asarray(qp)
energies = np.asarray(energies)
energies = energies - min(energies)
```

```python
plt.plot(qps, energies)
plt.show()
```
